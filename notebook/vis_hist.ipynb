# plot histogram of one input tif

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
def calculate_hist(i):
    frame = np.load(os.path.join(frame_dir,frames_files[i]))

    a = plt.figure(1)
    plt.title('frame0: mclean_roi_slope')
    frame0 = plt.hist(frame[:,:,0])
    b = plt.figure(2)
    plt.title('frame1: mclean_roi_aspect')
    frame1 = plt.hist(frame[:,:,1])
    c = plt.figure(3)
    plt.title('frame2: mclean_roi_rough')
    frame2 = plt.hist(frame[:,:,2])
    d = plt.figure(4)
    plt.title('frame3: mclean_roi_tpi')
    frame3 = plt.hist(frame[:,:,3])
    e = plt.figure(5)
    plt.title('frame4: mclean_roi_tri')
    frame4 = plt.hist(frame[:,:,4])
    f = plt.figure(6)
    plt.title('frame5: mclean_roi')
    frame5 = plt.hist(frame[:,:,5])
    print(np.min(frame[:,:,5]),np.max(frame[:,:,5]))
    plt.show()
```

```python
calculate_hist(1)
```
